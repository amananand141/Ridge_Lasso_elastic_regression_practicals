```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv', header=1)
```

```python
df.head()
```

```python
df.info()
```

## Data Cleaning

```python
## Missing values
df[df.isnull().any(axis=1)]
```

```python
### The dataset is divided into two regions
```

```python
df.loc[:122, "Region"]=0
df.loc[122:, "Region"]=1
```

```python
df
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## Remove the null values
df = df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## fix spaces in column names
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

### Change the required columns as integer data type

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

### Changing the other columns to float

```python
objects = [features for features in df.columns if df[features].dtypes=='str']
```

```python
objects
```

```python
for i in objects:
    # we're not changing the Classes column because it is a categorical feature and it needs to be object
    if i!='Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head(10)
```

```python
## Let's save the cleaned dataset
df.to_csv('Algerian_forest_fires_dataset_Cleaned.csv', index=False)
```
